# NFL Draft 2026 — Player Overview Report

High-level summary statistics and visualizations of the 2026 NFL Draft prospect pool.

```python
import sys
sys.path.insert(0, '../..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.data_loader import DATA_PROCESSED_DIR

sns.set_theme(style='whitegrid')
```

## 1. Load processed data

```python
df = pd.read_csv(DATA_PROCESSED_DIR / "players_clean.csv")
print(f"Total prospects: {len(df):,}")
df.head()
```

## 2. Prospects by position

```python
position_counts = df['position'].value_counts()

fig, ax = plt.subplots(figsize=(12, 5))
position_counts.plot(kind='bar', ax=ax, color='steelblue')
ax.set_title('NFL Draft 2026 — Prospects by Position')
ax.set_xlabel('Position')
ax.set_ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 3. Prospects by college

```python
top_colleges = df['college'].value_counts().head(15)

fig, ax = plt.subplots(figsize=(12, 5))
top_colleges.plot(kind='bar', ax=ax, color='darkorange')
ax.set_title('NFL Draft 2026 — Top 15 Colleges by Number of Prospects')
ax.set_xlabel('College')
ax.set_ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
