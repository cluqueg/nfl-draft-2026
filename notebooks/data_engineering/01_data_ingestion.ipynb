# NFL Draft 2026 — Data Ingestion

This notebook loads the raw CSV player data, applies basic cleaning and type casting,
and saves the result to `data/processed/` for use by reporting notebooks.

```python
import sys
sys.path.insert(0, '../..')

import pandas as pd
from src.data_loader import DATA_RAW_DIR, save_processed
```

## 1. Load raw data

Place your CSV file(s) in `data/raw/` and update the filename below.

```python
# Update this filename to match your uploaded CSV
RAW_FILE = "players.csv"

df = pd.read_csv(DATA_RAW_DIR / RAW_FILE)
print(f"Loaded {len(df):,} rows and {df.shape[1]} columns")
df.head()
```

## 2. Inspect & clean

```python
print(df.dtypes)
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Drop duplicate rows
df = df.drop_duplicates()

# Strip leading/trailing whitespace from string columns
str_cols = df.select_dtypes(include='object').columns
df[str_cols] = df[str_cols].apply(lambda col: col.str.strip())

print(f"Rows after deduplication: {len(df):,}")
```

## 3. Save processed data

```python
save_processed(df, "players_clean.csv")
print("Saved to data/processed/players_clean.csv")
```
